# RFM Analysis

```python
data.head()
```

```python
data.dtypes
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data['Customer ID'].nunique()
```

```python
data.dtypes
```

```python
max_date=data['Order Date'].max()
max_date
```

```python
dm = data[data['Order Date'] == data['Order Date'].max()]
dm
```

```python
recency=data.groupby('Customer ID')['Order Date'].max().reset_index()
recency.rename(columns={'Order Date':'max_order_date'},inplace=True)
recency
```

```python
recency['recency']=(data['Order Date'].max() - recency['max_order_date']).dt.days
```

```python
recency
```

```python
frequency=data.groupby('Customer ID')['Order ID'].nunique().reset_index()
frequency.rename(columns={'Order ID':'frequency'},inplace=True)
frequency
```

```python
monetary=data.groupby('Customer ID')['Profit'].sum().reset_index()
monetary.rename(columns={'Profit':'Monetary'},inplace=True)
monetary
```

```python
rfm=pd.merge(recency,frequency,on='Customer ID')
rfm
```

```python
rfm=pd.merge(rfm,monetary,on='Customer ID')
```

```python
rfm
```

```python
rfm['r_score']=pd.qcut(rfm['recency'],10,labels=range(10,0,-1))
rfm
```

```python
rfm['f_score']=pd.qcut(rfm['frequency'],10,labels=range(1,11))
rfm['m_score']=pd.qcut(rfm['Monetary'],10,labels=range(1,11))
rfm
```

```python
rfm['rfm_score'] = rfm['r_score'].astype(int) + rfm['f_score'].astype(int) + rfm['m_score'].astype(int)
rfm.rename(columns={'rfm_score':'rfm'},inplace=True)
rfm
```

```python
rfm['rfm_score']=pd.qcut(rfm['rfm'],10,labels=range(1,11))
#les clients avec le rfm le plus bas ayant 1 et les clients avec le rfm le plus haut auront 10
rfm
```
